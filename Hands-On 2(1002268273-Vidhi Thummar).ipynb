## INSERTION SORT

```python
def insertion(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

## SELECTION SORT

```python
def selection(arr):
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
```

## BUBBLE SORT

```python
def bubble(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

## BENCHMARK

```python
import random
import time
import matplotlib.pyplot as plt


def benchmark_sorting_algorithm(sort_function, arr):
    start_time = time.time()
    sort_function(arr)
    return time.time() - start_time


array_sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000]
times_insertion = []
times_selection = []
times_bubble = []

for size in array_sizes:
    arr = random.sample(range(1, 100000), size) 
   
    times_insertion.append(benchmark_sorting_algorithm(insertion, arr.copy()))
    times_selection.append(benchmark_sorting_algorithm(selection, arr.copy()))
    times_bubble.append(benchmark_sorting_algorithm(bubble, arr.copy()))

# Plot the benchmarking results
plt.plot(array_sizes, times_insertion, label="Insertion Sort", marker='o')
plt.plot(array_sizes, times_selection, label="Selection Sort", marker='x')
plt.plot(array_sizes, times_bubble, label="Bubble Sort", marker='s')

plt.xlabel('Array Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Benchmarking Sorting Algorithms')
plt.legend()
plt.grid(True)
plt.show()
```

## Selection Sort Correctness

Selection Sort is correct because:
1. **Invariant:** After each iteration of the outer loop, the smallest element in the unsorted portion of the array is placed at the beginning of the array.
2. **Loop Process:** The algorithm selects the smallest element from the unsorted portion and swaps it with the first element in the unsorted portion.
3. **Termination:** The process continues until all elements are sorted.

At the end of the algorithm, the entire array is sorted.
